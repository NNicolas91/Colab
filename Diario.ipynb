```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
def readDatosx(filename):
    #open file 
    file8 = open(filename, "r")
    data_read = file8.readlines()
    file8.close()
    
    #split data
    all_data = []
    
    for lineas in data_read:
        datos = lineas.split()
        all_data.append(datos)
    
    #convert data string to float x
    data = [] 
    a = np.shape(all_data)  
    
    for i in range(1, a[0]) :
        datos_float = np.asarray(all_data[i][0], dtype=np.float64)
        data.append(datos_float)
        
    return  data     
```

```python
def readDatosy(filename):
    #open file 
    file8 = open(filename, "r")
    data_read = file8.readlines()
    file8.close()
    
    #split data
    all_data = []
    
    for lineas in data_read:
        datos = lineas.split()
        all_data.append(datos)
    
    #convert data string to float y 
    data = [] 
    a = np.shape(all_data)  
    
    for i in range(1, a[0]) :
        datos_float = np.asarray(all_data[i][1], dtype=np.float64)
        data.append(datos_float)
        
    return  data 
```
